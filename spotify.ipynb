```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans

# Load dataset
dataset = pd.read_csv('spotify dataset.csv')

# Drop unnecessary columns (adjust column indexes based on your dataset)
dataset.drop(dataset.columns[[0, 1, 2, 3, -2]], axis=1, inplace=True)

# Define feature set
X = dataset.copy()

# One-Hot Encoding for categorical features (-7 and -4 columns, update as needed)
ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [-7, -4])], remainder='passthrough')
X_train = np.array(ct.fit_transform(X))

# Apply KMeans Clustering
km = KMeans(n_clusters=10, init="k-means++", random_state=5)
y_means = km.fit_predict(X_train)

# Add cluster labels to dataset
dataset["Cluster"] = y_means

# Compute correlation matrix (focusing only on 'track_popularity')
corr_df = dataset.corr(method="pearson")[['track_popularity']].sort_values(by='track_popularity', ascending=False)

# Plot heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(
    corr_df, 
    annot=True, 
    fmt=".2f",  # Show two decimal places
    cmap="inferno", 
    linewidths=1, 
    linecolor="black", 
    cbar=True,  
    annot_kws={"size": 12}  
)

# Formatting
plt.title("Correlation with Track Popularity", fontsize=14)
plt.xticks(fontsize=12, rotation=45, ha="right")
plt.yticks(fontsize=12, rotation=0)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select features for clustering
features = ["danceability", "energy", "loudness", "speechiness", "acousticness",
            "instrumentalness", "liveness", "valence", "tempo", "duration"]

# Standardize features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Determine optimal number of clusters using Elbow Method
inertia = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker="o", linestyle="-", color="teal")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```
